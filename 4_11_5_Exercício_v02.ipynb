```
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
```

```
ratings = pd.read_csv('ratings.csv')
movies = pd.read_csv('movies.csv')
```

```
movies = movies[['movieId', 'title']]
```

```
ratings_mean = ratings.groupby('movieId').mean().reset_index()

movies_rating = movies.merge(ratings_mean, on='movieId').sort_values(by='rating', ascending=False)
movies_rating.head(10)
```

```
# id don't like 559, 44912, 503
# id like 122917, 24428, 862

dont_like = [87430, 45442, 52722]
like = [1, 527, 89745]

dl = movies_rating[movies_rating['movieId'].isin(dont_like)].sort_values(by='rating')
l = movies_rating[movies_rating['movieId'].isin(like)].sort_values(by='rating')

print('\nLike:')
print(l)
print('\nDon\'t like:')
print(dl)
```

```
fig, ax = plt.subplots(dpi=180)

index = np.arange(3)

ax.bar(index, dl['rating'], width=0.4, color='IndianRed', label='Don\'t like')
ax.bar(index + 0.4, l['rating'], width=0.4, color='SkyBlue', label='Like')

ax.set_title('Notas de filmes', fontdict={'fontsize': 20})
ax.set_xlabel('Título')
ax.set_ylabel('Média das avaliações')
ax.legend(loc='lower right')
  
xticks = [0]

for i in np.arange(len(l)):
  xticks.append(dl.reset_index()['title'][i])
  xticks.append(l.reset_index()['title'][i])

ax.set_xticklabels(xticks, rotation=90)

plt.savefig('output.png')
plt.show()
```

```
fig, ax = plt.subplots(dpi=300)

index = np.arange(3)

dl_labels = '\n'.join(['L{index} {movie_name}'.format(index=idx, movie_name=i) for idx, i in enumerate(dl.reset_index()['title'])])
l_labels = '\n'.join(['DL{index} {movie_name}'.format(index=idx, movie_name=i) for idx, i in enumerate(l.reset_index()['title'])])

ax.bar(index, dl.reset_index()['rating'], width=0.4, color='IndianRed', label=dl_labels)
ax.bar(index+0.4, l.reset_index()['rating'], width=0.4, color='SkyBlue', label= l_labels)

ax.set_title('Notas de filmes', fontdict={'fontsize': 20})
ax.set_xlabel('Título')
ax.set_ylabel('Média das avaliações')
lgd = ax.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

xticks = [0]

for i in np.arange(len(l)):
  xticks.append(''.join(['L', np.str(i)]))
  xticks.append(''.join(['DL', np.str(i)]))

ax.set_xticklabels(xticks, rotation=30)

# fig.savefig('output.png', orientation='landscape', dpi=180)
fig.savefig('output', bbox_extra_artists=(lgd,), bbox_inches='tight')
plt.show()
```
