7. Fazer uma query exclusiva para o estado com maior número de relatos, buscando cidades que possuam um número superior a 10 relatórios. Enfatizar a cidade, a quantidade de relatos e formato do objeto não identificado.

```
states = df.groupby('State').sum().sort_values(by='Unnamed: 0', ascending=False)
states.columns = ['Quantidade']
best_state = states.index[0]
print('Estado com maior número de relatos: ', best_state)
```

```
city_freq_df = df[df['State'] == best_state].groupby('City').sum().sort_values(by='Unnamed: 0', ascending=False)
city_freq_df.columns = ['Quantidade']
city_freq_df = city_freq_df[city_freq_df['Quantidade'] > 10]
city_freq_greater_than_ten = city_freq_df.index
```

```
city_shape_df = df[df['City'].isin(city_freq_greater_than_ten)].groupby(['City', 'Shape']).sum()
city_shape_df.columns = ['Quantidade']
city_shape_df
```
