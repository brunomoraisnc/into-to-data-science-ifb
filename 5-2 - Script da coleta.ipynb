```
import numpy as np
import pandas as pd
import requests

from pprint import pprint
```

```
df = pd.read_html('http://www.nuforc.org/webreports/ndxevent.html')
```

```
df = df[0].copy()
```

```
sep_2017_idx = df[df['Reports'] == '09/2017'].index[0]
sep_1997_idx = df[df['Reports'] == '09/1997'].index[0]+1
```

```
df = df.iloc[sep_2017_idx:sep_1997_idx]
df.reset_index(inplace=True)
```

```
df
```

```
# df['Reports'].str.split("/").apply(lambda x: 'ndxe' + ''.join([x[1],x[0]]) + '.html')
df['Reports'] = df['Reports'].str.split("/").apply(lambda x: ''.join([x[1],x[0]]))
df.head()
```

```
base_url = 'http://www.nuforc.org/webreports/ndxe' 
urls = base_url + df['Reports'] + '.html'
urls.tail()
```

```
# df2 = pd.read_html(urls[0])[0]
df2 = pd.DataFrame()
it = 0

for i in urls:
  new_df = pd.read_html(i)[0]
  df2 = pd.concat([df2, new_df], ignore_index=True)
  it = it +1
  print(it)
  
df2
```

```
df2.to_csv('ovni_dataset.csv')
```
